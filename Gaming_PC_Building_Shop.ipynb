```python
# Gaming PC Building Shop - CRUD Project

class Component:
    def __init__(self, name, category, brand, price, stock):
        self.name = name
        self.category = category  # CPU, GPU, RAM, etc.
        self.brand = brand
        self.price = price
        self.stock = stock

    def __str__(self):
        return (f"ID: {self.id} | Name: {self.name} | Category: {self.category} | "
                f"Brand: {self.brand} | Price: ‚Çπ{self.price} | Stock: {self.stock} units")


class PCShop:
    def __init__(self):
        self.components = {}
        self.next_id = 1  # Unique ID for each component

    # CREATE
    def add_component(self, name, category, brand, price, stock):
        comp = Component(name, category, brand, price, stock)
        comp.id = self.next_id
        self.components[self.next_id] = comp
        self.next_id += 1
        return f"‚úÖ Component '{comp.name}' (ID: {comp.id}) added successfully!"

    #READ (All)
    def view_all_components(self):
        if not self.components:
            return "No components available in the shop."
        for comp in self.components.values():
            print(comp)

    #READ (Single)
    def view_component(self, comp_id):
        comp = self.components.get(comp_id)
        if comp:
            return comp
        else:
            return f"‚ùå Component with ID {comp_id} not found."

    #UPDATE
    def update_component(self, comp_id, name=None, category=None, brand=None, price=None, stock=None):
        comp = self.components.get(comp_id)
        if comp:
            if name:
                comp.name = name
            if category:
                comp.category = category
            if brand:
                comp.brand = brand
            if price is not None:
                comp.price = price
            if stock is not None:
                comp.stock = stock
            return f"üõ†Ô∏è Component {comp_id} updated successfully!"
        else:
            return f"‚ùå Component with ID {comp_id} not found."

    #DELETE
    def delete_component(self, comp_id):
        if comp_id in self.components:
            del self.components[comp_id]
            return f"üóëÔ∏è Component {comp_id} deleted successfully!"
        else:
            return f"‚ùå Component with ID {comp_id} not found."


#Testing the PC Shop Management System

shop = PCShop()

# Create components
print(shop.add_component("Ryzen 7 5800X", "CPU", "AMD", 28000, 10))
print(shop.add_component("RTX 4070 Ti", "GPU", "NVIDIA", 78000, 5))
print(shop.add_component("Corsair Vengeance 16GB", "RAM", "Corsair", 6500, 25))

print("\nüßæ All Available Components:")
shop.view_all_components()

# Read single component
print("\nüîç Viewing Component ID 2:")
print(shop.view_component(2))

# Update
print("\n‚úèÔ∏è Updating Component ID 3:")
print(shop.update_component(3, price=6200, stock=30))

# Delete
print("\n‚ùå Deleting Component ID 1:")
print(shop.delete_component(1))

print("\nüì¶ Updated List of Components:")
shop.view_all_components()
```
