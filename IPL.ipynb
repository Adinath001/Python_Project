```python
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns

# Load your dataset
df = pd.read_csv("ODI data.csv")

# Show first few rows
print(df.head())
```

```python
top_runs = df.sort_values('Runs', ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.bar(top_runs['Player'], top_runs['Runs'], color='gold')
plt.title("Top 10 Run Scorers in ODI")
plt.xlabel("Player")
plt.ylabel("Runs")
plt.xticks(rotation=45)
plt.show()
```

```python
top5 = df.sort_values('Runs', ascending=False).head(5)
plt.bar(top5['Player'], top5['Runs'])
plt.title("Top 5 Run Scorers")
plt.xlabel("Player")
plt.ylabel("Runs")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("ODI data.csv")
print(df.head())
print(df.columns)
```

```python
plt.plot(top5['Player'], top5['Runs'], marker='o', label='Runs')
plt.plot(top5['Player'], top5['Mat'], marker='s', label='Matches')

plt.title("Top 5 Players: Runs vs Matches")
plt.xlabel("Player")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 1️⃣ Bar chart – Top 10 players by Runs
plt.bar(df['Player'].head(10), df['Runs'].head(10))
plt.title("Top 10 Players by Runs")
plt.xticks(rotation=45)
plt.ylabel("Runs")
plt.show()
```

```python
# 2️⃣ Bar chart – Top 10 players by Batting Average
plt.bar(df['Player'].head(10), df['Ave'].head(10), color='orange')
plt.title("Top 10 Players by Average")
plt.xticks(rotation=45)
plt.ylabel("Average")
plt.show()
```

```python
# 3️⃣ Line chart – Runs vs Balls Faced (BF)
top5 = df.head(5)
plt.plot(top5['Player'], top5['Runs'], marker='o', label='Runs')
plt.plot(top5['Player'], top5['BF'], marker='s', label='Balls Faced')
plt.title("Runs vs Balls Faced")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# 4️⃣ Scatter plot – Runs vs Strike Rate
plt.scatter(df['Runs'], df['SR'], color='green')
plt.title("Runs vs Strike Rate")
plt.xlabel("Runs")
plt.ylabel("Strike Rate")
plt.show()
```

```python
# 5️⃣ Histogram – Distribution of Batting Average
plt.hist(df['Ave'].dropna(), bins=10, color='purple', edgecolor='black')
plt.title("Distribution of Batting Average")
plt.xlabel("Average")
plt.ylabel("Frequency")
plt.show()
```

```python
# 6️⃣ Pie chart – 100s among top 5 players
top5 = df.head(5)
plt.pie(top5['100'], labels=top5['Player'], autopct='%1.1f%%', startangle=140)
plt.title("Century (100s) Contribution of Top 5 Players")
plt.show()
```

```python
# 7️⃣ Horizontal Bar chart – Half-centuries (50s)
plt.barh(df['Player'].head(10), df['50'].head(10), color='skyblue')
plt.title("Top 10 Players by 50s")
plt.xlabel("Number of 50s")
plt.ylabel("Player")
plt.show()
```

```python
# 8️⃣ Box plot – Distribution of Strike Rate
# Convert SR column to numeric (force errors to NaN)
df['SR'] = pd.to_numeric(df['SR'], errors='coerce')

# Drop missing values and plot
plt.boxplot(df['SR'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightblue'))

plt.title("Boxplot of Strike Rate")
plt.ylabel("Strike Rate")
plt.show()
```

```python
# 9️⃣ Line chart – Runs vs Matches (Top 5)
top5 = df.head(5)
plt.plot(top5['Mat'], top5['Runs'], marker='o')
plt.title("Runs vs Matches (Top 5 Players)")
plt.xlabel("Matches")
plt.ylabel("Runs")
plt.show()
```

```python
# 🔟 Bar chart – Not Outs (NO) among Top 10 Players
plt.bar(df['Player'].head(10), df['NO'].head(10), color='teal')
plt.title("Top 10 Players by Not Outs (NO)")
plt.xticks(rotation=45)
plt.ylabel("Not Outs")
plt.show()
```
